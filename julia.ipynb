```julia
using Parquet
using DataFrames
```

```julia
function download_taxi_data(month, year)::DataFrame
    url = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_$(year)-$(month).parquet"
    filename = "green_tripdata_$(year)-$(month).parquet"
    download(url, filename)
    df = DataFrame(read_parquet(filename))
    return df
end
```

```julia
download_taxi_data("01", 2019)
```
