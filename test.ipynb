```python
import polars as pl
```

```python
def read_taxi_data_from_tlc(year, month, colour) -> pl.DataFrame:
    """The main ETL function"""
    dataset_file = f"tripdata_{year}-{month:02}"
    dataset_url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/{colour}_{dataset_file}.parquet"
    data = pl.read_parquet(dataset_url)
    return data
```

```python
data = read_taxi_data_from_tlc(2019, 1, "yellow")
```

```python
data.with_columns((pl.col("tpep_dropoff_datetime") - pl.col("tpep_pickup_datetime")).alias("trip_duration"))
```
